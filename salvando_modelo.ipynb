```
# importaçães
import pickle
import pandas as pd
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```
# Carregando o data set
dataset = pd.read_csv('Iris.csv')
```

```
dataset.head()
```

```
# separando a os dados e realizando tratamento

# separando base de dados em previsores(X) e classificadores(Y)
X = pd.DataFrame(dataset.iloc[:,1:-1])
Y = pd.DataFrame(dataset.iloc[:, -1])

# encondando o classificador para poder ser usado nos modelos de ia
encoder_y = LabelEncoder() 
Y = pd.DataFrame( encoder_y.fit_transform(Y))
```

```
# criando base de treinamento e teste
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42 )
```

```
# criando classificador svm

classSVM = SVC(kernel = 'rbf', C = 2.0)
classSVM.fit(X, Y)
```

```
# criando classificador random forest
classRandomForest = RandomForestClassifier(n_estimators = 40, criterion = 'entropy')
classRandomForest.fit(X, Y)
```

```
# criando classificador de regressão linear
classLinearRegression = LinearRegression()
classLinearRegression.fit(X,Y)
```

```
# salvando os modelos em um arquivo .sav com pickle
pickle.dump(classLinearRegression, open('modelo_regressao_linear.sav', 'wb'))
pickle.dump(classSVM, open('modelo_svm.sav', 'wb'))
pickle.dump(classRandomForest, open('modelo_random_forest.sav', 'wb'))
```

Com isso salvamos em um arquivo os modelos treinados

```
# avaliando o score dos modelos gerados
resultado_classSVM = classSVM.score(X, Y)
resultado_classRandomForest = classRandomForest.score(X, Y)
resultado_classLinearRegression = classLinearRegression.score(X,Y)
```
